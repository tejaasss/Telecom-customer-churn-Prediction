```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r'C:\Users\Tejas\Jupyter Notebooks\Projects\MLProject-ChurnPrediction-main\tele_churn.csv')
```

```python
df.drop('Unnamed: 0', inplace=True, axis=1)
```

```python
df.columns
```

```python
X = df.drop('Churn', axis=1)
```

```python
X.shape
```

```python
y = df['Churn']
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
```

```python
params = {'criterion':['gini','entropy'], 'max_depth':np.arange(3,15), 'min_samples_leaf': np.arange(4,20,2)}
dtc = DecisionTreeClassifier()
```

```python
model = GridSearchCV(dtc, params, cv=5, verbose=2)
```

```python
model.fit(X_train, y_train)
```

```python
print(model.best_estimator_)
print(model.best_params_)
print(model.best_score_)
```

```python
dtc_tuned = DecisionTreeClassifier(criterion='gini', max_depth=5, min_samples_leaf=18)
```

```python
dtc_tuned.fit(X_train, y_train)
```

```python
y_pred = dtc_tuned.predict(X_test)
```

```python
import sys
np.set_printoptions(threshold=sys.maxsize)
y_pred
```

```python
y_test = y_test.to_numpy()
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

```python
print(confusion_matrix(y_test, y_pred))
```

```python
print(accuracy_score(y_test, y_pred))
```

```python
print(classification_report(y_test, y_pred, labels=[0,1]))
```

```python
#upsampling using smoteenn
from imblearn.combine import SMOTEENN
sm = SMOTEENN()
```

```python
X_resampled, y_resampled = sm.fit_resample(X,y)
```

```python
Xr_train, Xr_test, yr_train, yr_test = train_test_split(X_resampled, y_resampled, test_size=0.2, random_state=0)
```

```python
dtc_resampled = DecisionTreeClassifier(criterion='entropy', max_depth=6, min_samples_leaf=8)
```

```python
dtc_resampled.fit(Xr_train, yr_train)
```

```python
yr_pred = dtc_resampled.predict(Xr_test)
```

```python
print(confusion_matrix(yr_test, yr_pred))
```

```python
print(accuracy_score(yr_test, yr_pred))
```

```python
print(dtc_resampled.score(Xr_test, yr_test))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf_model = RandomForestClassifier(n_estimators = 100, criterion='entropy', max_depth=6, min_samples_leaf=8)
```

```python
rf_model.fit(X_train, y_train)
```

```python
y_pred = rf_model.predict(X_test)
```

```python
print(classification_report(y_test, y_pred))
```

```python
#upsampling with randomForestClassifier
rf_resampled = RandomForestClassifier(n_estimators = 100, criterion='gini', max_depth=6, min_samples_leaf=8)
rf_resampled.fit(Xr_train, yr_train)
```

```python
yr_rf_pred = rf_resampled.predict(Xr_test)
```

```python
print(accuracy_score(yr_rf_pred, yr_test))
print(classification_report(yr_test, yr_rf_pred))
```

```python
print(confusion_matrix(yr_test, yr_rf_pred))
```

```python
#trying principal component analysis
from sklearn.decomposition import PCA
pca = PCA(0.9)
Xr_train_pca = pca.fit_transform(Xr_train)
Xr_test_pca = pca.transform(Xr_test)
explained_variance = pca.explained_variance_ratio_
```

```python
pca_rf_model=RandomForestClassifier(n_estimators=100, criterion='gini', random_state = 100,max_depth=6, min_samples_leaf=8)
```

```python
pca_rf_model.fit(Xr_train_pca, yr_train)
yr_pred_pca = pca_rf_model.predict(Xr_test_pca)
```

```python
print(accuracy_score(yr_pred_pca, yr_test))
print(classification_report(yr_test, yr_pred_pca))
```
